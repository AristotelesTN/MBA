```

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn.naive_bayes import GaussianNB

%matplotlib inline
```

```
# ATIVIDADE DA AULA 1 - MBA MACHINE LEARNING (UTILIZANDO KMEANS COMO METODO N√ÉO SUPERVISIONADO E GUASSIAN NAIVE BAYES COMO METODO SUPERVISIONADO)
```

```
# CRIANDO DATASETS
x, y = make_blobs(n_samples=160, n_features=4, centers=8, random_state=1234, cluster_std=0.8)
x1, y1 = make_blobs(n_samples=160, n_features=4, centers=8, random_state=0, cluster_std=0.5)


print (x.shape)
print (y.shape) # LABELS 1

print (x1.shape)
print (y1.shape) # LABELS 2
```

```
plt.figure(figsize=(6, 6))
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.title('CLUSTERS 1')


plt.figure(figsize=(6,6))
plt.scatter(x1[:, 0], x1[:, 1], c=y1)
plt.title('CLUSTERS 2')
```

```
#KMEANS ATRAV√âS DO SCIKIT 

kmeans = KMeans(n_clusters=8, max_iter=500, random_state=1234)
kmeans.fit(x)

pred = kmeans.predict(x)
```

```
#print(kmeans.cluster_centers_)

#print(kmeans.labels_)

#print(pred)

plt.figure(figsize=(6, 6))
plt.scatter(x[:,0], x[:,1],c=pred)
#plt.scatter(x1[:,0], x1[:,1],c=pred)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=75, c='red')

```

```
#GUASSIAN NAIVE BAYES ATRAV√âS DO SCIKIT 
clf = GaussianNB()
clf.fit(x, y)
```

```
#UTILIZANDO CLUSTER 1 PARA PREDIZER A CLASSIFICA√á√ÉO

y_pred = clf.predict(x)

print(clf.score(x,y))

plt.figure(figsize=(6, 6))
plt.scatter(x[:,0], x[:,1],c=y_pred)
```

```
#UTILIZANDO CLUSTER 2 PARA PREDIZER A CLASSIFICA√á√ÉO
y1_pred = clf.predict(x1)

print(clf.score(x1,y))

plt.figure(figsize=(6, 6))
plt.scatter(x1[:,0], x1[:,1],c=y1_pred)
```
